```python
import random

# 問題（日本語 : カタルーニャ語）
words = {
    "懐中電灯": "llanterna",
    "マッチ箱": "capsa de llumins",
    "食堂": "cantimplora",
    "多目的ナイフ": "navalla multiús",
    "コンパス": "brúixola",
    "薬箱":"farmaciola",
    "虫よけスプレー":"repel·lent de mosquitos",
    "コーチ":"autocar",
    "カード":"corda",
    "ハサミ":"tisores",
    "アウトドアウェア":"anorac"
}

# ランダムで問題を出す
japanese = random.choice(list(words.keys()))
answer = words[japanese]

print("次の単語をカタルーニャ語で答えてください")
print("👉", japanese)

user = input("答え：")

if user.strip().lower() == answer.lower():
    print("正解！🎉")
else:
    print("不正解")
    print("正しくは：", answer)
```
